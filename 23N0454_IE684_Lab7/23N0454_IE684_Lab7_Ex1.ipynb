Note that if $yw^Tx = z$ then, for the losses, $L_h, L_l, L_{sh}$ we get,

To express the loss functions $G_h(z), G_l(z)$, and $G_{s h}(z)$ as functions of $z$, we can substitute $z=$ $y w^T x$ into the original loss functions:
1. For the hinge loss $L_h\left(y, w^T x\right)=\max \left(0,1-y w^T x\right)$ :
$$
G_h(z)=\max (0,1-z)
$$
2. For the logistic loss $L_l\left(y, w^T x\right)=\log \left(1+\exp \left(-y w^T x\right)\right)$ :
$$
G_l(z)=\log (1+\exp (-z))
$$
3. For the squared hinge loss $L_{s h}\left(y, w^T x\right)=\left(\max \left(0,1-y w^T x\right)\right)^2$ :
$$
G_{s h}(z)=(\max (0,1-z))^2
$$

```python
# Define loss functions

def hinge_loss(z):
    return np.maximum(0, 1 - z)

def logistic_loss(z):
    return np.log(1 + np.exp(-z))

def squared_hinge_loss(z):
    return (np.maximum(0, 1 - z))**2

# Generate values for z, I have taken z from -2.5 to 2.5 since I cannot take whole real line
z_values = np.linspace(-2.5, 2.5, 1000)

# Calculate loss values
hinge_losses = hinge_loss(z_values)
logistic_losses = logistic_loss(z_values)
squared_hinge_losses = squared_hinge_loss(z_values)

# Plotting
plt.plot(z_values, hinge_losses, label='Hinge Loss', color='blue')
plt.plot(z_values, logistic_losses, label='Logistic Loss', color='red')
plt.plot(z_values, squared_hinge_losses, label='Squared Hinge Loss', color='green')

# Add labels and legend
plt.xlabel('z')
plt.ylabel('Loss')
plt.title('Comparison of Loss Functions')
plt.legend()
plt.grid(True)
plt.show()
```

In the plot, the behavior of the respective loss functions with respect to z is as follows:

Hinge Loss $G_h(z)$: It remains constant at 0 until z becomes less than 1, after which it linearly increases with z. It penalizes misclassifications linearly. It alsos show non-differentiability at point x = 1 which will play a role in descent methods.

Logistic Loss $G_l(z)$: It smoothly increases as z moves towards negative infinity, reaching 0 as z approaches positive infinity. It penalizes misclassifications more severely as z moves away from 0. The curve is smooth and differentiable on real line allowing to use different gradient based methods efficiently.

Squared Hinge Loss $G_{sh}(z)$: It behaves similarly to the hinge loss but penalizes misclassifications quadratically. This results in a smoother curve compared to the hinge loss. It also increases much faster than logistic loss for $z \leq -2$.
