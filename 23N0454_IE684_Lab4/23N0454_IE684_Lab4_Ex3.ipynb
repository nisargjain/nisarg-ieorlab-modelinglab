#Exercise 3.1: Convergence of sequences:#

Q0.png

```python
tolerance = 1e-8


def seq1(k):
  return (1+0.005)**(-(2^k))

prev = seq1(0)
current = seq1(1)
k = 2
while abs(current - prev) > tolerance:
  prev= current
  current = seq1(k)
  k += 1

print(f'Converged seq for 1 is x* = {seq1(k)} and time for convergence k = {k}')
```

```python
tolerance = 1e-8


def seq2(k):
  return (1+0.005**k)

prev = seq2(0)
current = seq2(1)
k = 2
while abs(current - prev) > tolerance:
  prev= current
  current = seq2(k)
  k += 1

print(f'Converged seq for 2 is x* = {seq2(k)} and time for convergence k = {k}')
```

```python
tolerance = 1e-8

def seq3(k):
  return (1+(0.005**(-k)))

prev = seq3(0)
current = seq3(1)
k = 2
while abs(current - prev) > tolerance:
  prev= current
  current = seq3(k)
  k += 1

print(f'Converged seq for 3 is x* = {seq3(k)} and time for convergence k = {k}')
```

which means sequence is diverging

```python
tolerance = 1e-8

def seq4(k):
  return (1+(k**(-k)))

for k in range(2,100):
  print(seq4(k))
```
