```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma
from matplotlib import ticker, cm
from math import sqrt, pi
from numpy import exp, cos, sin
```

```python
def f1(x,y):
  return (100*(y**2)) + (100*(x**4)) - (200*y*(x**2)) + (x**2) - (2*x) + 1

def f2(x,y):
  return x**4 + y**4 - (20*x**3) -(20*y**3) + (100*x**2) + (100*y**2)

def f3(x,y):
  return -(cos(x)*cos(y)*exp(-((x-pi)**2 + (y-pi)**2)))

def f4(x,y):
  return (2*x**2) - (1.05*x**4) + (x**6)/6 + (x*y) + (y**2)
```

1.

```python
n = 100
x = np.linspace(-5, 5, n)
y = np.linspace(-5, 5, n)
X, Y = np.meshgrid(x, y)
Z = f1(X,Y)
p = plt.contourf(X,Y,Z,locator = ticker.LogLocator())
cbar = plt.colorbar(p)
plt.show()
```

2.

```python
n = 300
x = np.linspace(-10, 20, n)
y = np.linspace(-10, 20, n)
X, Y = np.meshgrid(x, y)
Z = f2(X,Y)
p = plt.contourf(X,Y,Z,locator = ticker.LogLocator())
# p = plt.contourf(X,Y,Z)
cbar = plt.colorbar(p)
plt.show()
```

```python
n = 300
x = np.linspace(1.5, 4.8, n)
y = np.linspace(1.5, 4.8, n)
X, Y = np.meshgrid(x, y)
Z = f3(X,Y)
p = plt.contourf(X,Y,Z)
cbar = plt.colorbar(p)
plt.show()
```

```python
n = 300
x = np.linspace(-10, 10, n)
y = np.linspace(-10, 10, n)
X, Y = np.meshgrid(x, y)
Z = f4(X,Y)
p = plt.contourf(X,Y,Z,locator = ticker.LogLocator())
cbar = plt.colorbar(p)
plt.show()
```
