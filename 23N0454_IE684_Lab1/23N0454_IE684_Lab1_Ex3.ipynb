this means function is $f(x,y) = (1.4-x)^2 + 60*(y-x^2)^2$

```python
x0 = (-1,1)
alpha = 0.001
taus = [10**(-p) for p in range(1,7)]

#gradient descent
from numpy import linalg as l2

def f3(x,y):
  return (1.4-x)**2 + 60*(y-x**2)**+2

def grad(x,y):
  return np.array([-360*x*(y-x**2) + 2*x - 2.8, 200*(y-x**2)])

minimizerstore = []
minimumstore = []
iterationstore = []

for tau in taus:
  i,j = x0[0], x0[1]
  count = 0
  while l2.norm(grad(i,j)) > tau:
    g_i, g_j = grad(i,j)
    i -= alpha*g_i
    j -= alpha*g_j
    count += 1
  minimizerstore.append([i,j])
  minimumstore.append(f3(i,j))
  iterationstore.append(count)
```
